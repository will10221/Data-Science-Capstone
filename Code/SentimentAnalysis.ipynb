List of tickers

```python
tickers = ["GOOG","BRKB","NVDA",'JPM','JNJ','META','V','HD','PG','MA','CVX','MRK','ABBV','LLY','BAC','AVGO',
          'PEP','PFE','KO','COST','TMO','WMT','MCD','DIS','ABT','WFC','ACN','CMCSA',
           'ADBE','CRM','DHR','VZ','TXN','LIN','NKE','NFLX','PM','BMY','NEE','RTX','QCOM','T',"CSCO",'XOM','UNH' 'AAPL','MSFT','AMZN'  
```

```python
#Output path is where the program will write files too
from google.colab import drive
drive.mount('/content/drive')
output_dir_path = "/content/drive/My Drive/Data_Capstone_Group2P/Data/Tweets/"
```

```python
#For loop that reads in every CSV of tweets into a pandas data frame
for ticker in tickers:
  df = pd.read_csv(ticker+"_tweets.csv")
  labels = {0:'neutral', 1:'positive',2:'negative'}
  sentiment = []

#For loop that inputs the tweets and prints the sentiment value
  for x in df['Tweet']:
        inputs = tokenizer(x, return_tensors="pt", padding=True)
        outputs = finbert(**inputs)[0]
        val = labels[np.argmax(outputs.detach().numpy())]
        #val = np.argmax(outputs.detach().numpy())
        sentiment.append(val)
  df['sentiment'] = sentiment
#Outputs the dataframe to output_dir_path 
  df.to_csv(output_dir_path + ticker+"_tweets.csv")
```
