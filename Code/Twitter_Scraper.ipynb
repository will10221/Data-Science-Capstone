```python
#Import libraries 
import pandas as pd
import snscrape.modules.twitter as sntwitter
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from transformers import BertTokenizer, BertForSequenceClassification
from datetime import datetime, date, timedelta

import warnings
%matplotlib inline
from datetime import date
```

```python
#Initialize the finbert and tokenizer models
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
```

```python
#Initialize list of tickers 
tickers = ['AAPL','MSFT','AMZN','GOOGL','BRK.B','NVDA','TSLA','GOOG','XOM','UNH',
           'JPM','JNJ','META','V','HD','PG','MA','CVX','MRK','ABBV','LLY','BAC','AVGO',
           'PEP','PFE','KO','COST','TMO','CSCO','WMT','MCD','DIS','ABT','WFC','ACN','CMCSA',
           'ADBE','CRM','DHR','VZ','TXN','LIN','NKE','NFLX','PM','BMY','NEE','RTX','QCOM','T']
```

```python
warnings.filterwarnings('ignore')
#Create list tweets
tweets=[]
#Loop through and add tweets to the list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('from:Microsoft').get_items()):
    if i > 20:
        break
    else:
        tweets.append([tweet.date, tweet.user.username,
                      tweet.content, tweet.likeCount, tweet.retweetCount])
#Add tweets to a dataframe
df = pd.DataFrame(tweets, columns = ['Date','username','tweet','num_of_likes',
                                    'num_of_retweet'])
```
